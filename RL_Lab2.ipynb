```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_arms = 5
true_reward_probs = [0.9, 0.8, 0.7, 0.6, 0.5]
```

```python
epsilon = 0.2
estimated_reward = np.zeros(num_arms)
num_pulls = np.zeros(num_arms)
total_reward = 0
reward_history = []
```

```python
for i in range(200):
  if np.random.random() < epsilon:
    chosen_arm = np.random.choice(num_arms)
  else:
    chosen_arm = np.argmax(estimated_reward)

    # Simulate pulling the chosen arm
  reward = np.random.binomial(n=1, p=true_reward_probs[chosen_arm])

    # Update variables
  total_reward += reward
  reward_history.append(total_reward / (i + 1))
  num_pulls[chosen_arm] += 1
  estimated_reward[chosen_arm] += (reward - estimated_reward[chosen_arm]) / num_pulls[chosen_arm]
```

```python
for arm in range(num_arms):
  print(f"Estimated reward probability for Arm {arm+1}: {estimated_reward[arm]:.4f}")
```

```python
plt.plot(reward_history)
plt.xlabel("Iterations")
plt.ylabel("Average Reward")
plt.title("Average Reward over Iterations")
plt.show()
```
