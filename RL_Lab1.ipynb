```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

# Generate random data
random.seed(42)

x_positive = [random.randint(1, 10) for _ in range(50)]
y_positive = [x + random.random() * 3 for x in x_positive]

x_negative = [random.randint(1, 10) for _ in range(50)]
y_negative = [-x + random.random() * 3 for x in x_negative]

x_zero = [random.randint(1, 10) for _ in range(50)]
y_zero = [random.random() * 10 for _ in range(50)]

# Create a DataFrame
data = {
    'x_positive': x_positive,
    'y_positive': y_positive,
    'x_negative': x_negative,
    'y_negative': y_negative,
    'x_zero': x_zero,
    'y_zero': y_zero
}

df = pd.DataFrame(data)

# Calculate correlations
corr_positive = df['x_positive'].corr(df['y_positive'])
corr_negative = df['x_negative'].corr(df['y_negative'])
corr_zero = df['x_zero'].corr(df['y_zero'])

# Plot scatter plots
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(df['x_positive'], df['y_positive'], color='blue')
plt.title(f"Positive Correlation\nCorrelation coefficient: {corr_positive:.2f}")

plt.subplot(1, 3, 2)
plt.scatter(df['x_negative'], df['y_negative'], color='red')
plt.title(f"Negative Correlation\nCorrelation coefficient: {corr_negative:.2f}")

plt.subplot(1, 3, 3)
plt.scatter(df['x_zero'], df['y_zero'], color='green')
plt.title(f"Zero Correlation\nCorrelation coefficient: {corr_zero:.2f}")
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({'a': np.random.randint(0, 50, 1000)})
df['b'] = df['a'] + np.random.normal(0, 10, 1000) # positively correlated with 'a'
df['c'] = 100 - df['a'] + np.random.normal(0, 5, 1000) # negatively correlated with 'a'
df['d'] = np.random.randint(0, 50, 1000) # not correlated with 'a'

df.corr()
```

```python
pd.plotting.scatter_matrix(df, figsize=(6, 6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```
